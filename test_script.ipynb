```python
from PIL import Image
import base64
import io

#with Image.open("mask.jpg") as im:

#Read image in as bytes IO
#img = io.BytesIO(bytes)


img = Image.open("mask.jpg")

#Conversion from image to bytes
bytes = img.tobytes()

#Encode the image bytes using base 64 and then decode to string
mystr = base64.b64encode(bytes).decode() 
#return as a string: AHB1AHF2AHN3AHV5AHd9AXh+AXqCAnuEBYCIBICKBICMAoCMAo......
```

```python
#Re-create the image from the byte string
newbytes = base64.b64decode(mystr)
'''
# Not enough data for A channel for the RGBA, use RGB instead
image = Image.frombytes("RGBA", (img.size), newbytes)
'''
image = Image.frombytes("RGB", (img.size), newbytes)
image.show()
```
